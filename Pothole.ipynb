```python
import yaml

yaml_path = "/root/.cache/kagglehub/datasets/anggadwisunarto/potholes-detection-yolov8/versions/3/data.yaml"

with open(yaml_path, "r") as f:
    data = yaml.safe_load(f)

# Fix paths
base_path = "/root/.cache/kagglehub/datasets/anggadwisunarto/potholes-detection-yolov8/versions/3"

data["train"] = base_path + "/train/images"
data["val"] = base_path + "/valid/images"

# Save locally
with open("data.yaml", "w") as f:
    yaml.dump(data, f)

print("✅ data.yaml fixed!")
```

```python
from ultralytics import YOLO

model = YOLO("yolov8m.pt")  # high accuracy

model.train(
    data="data.yaml",
    epochs=100,
    imgsz=640,
    batch=16,
    device=0
)
```

```python
from google.colab import files
files.download('/content/runs/detect/train/weights/best.pt')
```

```python
from ultralytics import YOLO
import matplotlib.pyplot as plt

model = YOLO("runs/detect/train/weights/best.pt")

# Run validation (this generates confusion matrix)
metrics = model.val()

# Save confusion matrix
cm = metrics.confusion_matrix.matrix

# Plot and save
plt.figure(figsize=(6,6))
plt.imshow(cm, cmap='Blues')
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")

for i in range(len(cm)):
    for j in range(len(cm)):
        plt.text(j, i, int(cm[i][j]), ha='center', va='center')

plt.colorbar()
plt.savefig("confusion_matrix.png")
plt.close()
```

```python
results = metrics.results_dict

with open("metrics.txt", "w") as f:
    for key, value in results.items():
        f.write(f"{key}: {value}\n")
```

```python
import numpy as np

# Example (use your actual arrays)
conf = metrics.curves[0][0]   # confidence values
recall = metrics.curves[0][1] # recall values

np.save("confidence.npy", conf)
np.save("recall.npy", recall)
```

```python
plt.plot(conf, recall)
plt.xlabel("Confidence")
plt.ylabel("Recall")
plt.title("PR Curve")
plt.savefig("pr_curve.png")
plt.close()
```
